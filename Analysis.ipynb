```python
import numpy as np
import pandas as pd
```

```python
names1880 = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/babynames/yob1880.txt', 
                        names=['name','sex','birth'])
names1880.head(10)
```

```python
names1880.groupby('sex').birth.sum()
```

```python
years = range(1880, 2011)
pieces = []
columns = ['names', 'sex', 'births']
```

```python
for year in years:
    path = 'https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/babynames/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)
```

```python
pieces
```

```python
names = pd.concat(pieces, ignore_index=True)
names
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
total_births.head()
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    group['prop'] = group.births/group.births.sum()
    return group
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
names
```

```python
names.groupby(['year','sex']).prop.sum()
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
#Drop the group index, not needed
top1000.reset_index(inplace=True, drop=True)
```

```python
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births', index='year', columns='names', aggfunc=sum)
total_births.info()
```

```python
subset = total_births[['John','Harry','Mary','Marilyn']]
subset.plot(subplots=True, figsize=(12,18), grid=False, title='Number of births per year')
```

```python
table = top1000.pivot_table('prop', index='year', columns='sex', aggfunc=sum)
table.plot(title='Sum of table1000.prop by year and sex', yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2020, 10))
```

```python
get_last_letter = lambda x: x[-1]
last_letters = names.names.map(get_last_letter)
last_letters.name = 'last_letter'
table = names.pivot_table('births', index=last_letters, columns=['sex','year'], aggfunc = sum)
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable/subtable.sum()
letter_prop
```

```python
import matplotlib.pyplot as plt
fix, axes = plt.subplots(2, 1, figsize=(10,8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female', legend=False)
```
